# Logistic Regression

Split the data into training set and testing test

```python
train.drop(['Ad Topic Line', 'City', 'Country', 'weekday', 'Timestamp'], axis=1, inplace=True)
```

```python
train.head()
```

```python
X = train.drop('Clicked on Ad', axis=1)
y = train['Clicked on Ad']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
log_regr = LogisticRegression()
```

```python
log_regr.fit(X_train, y_train)
```

# Predictions and Evaluations

```python
prediction = log_regr.predict(X_test)
```

```python
print(classification_report(y_test, prediction))
```

```python
confusion_matrix(y_test, prediction)
```

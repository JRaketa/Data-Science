```python
from sklearn.cluster import KMeans
```

```python
kmean = KMeans(n_clusters=2)
```

```python
df.columns
```

```python
X = df.drop('Private', axis=1)
categ_cols = ['Unnamed: 0']
final_data = pd.get_dummies(X, columns=categ_cols, drop_first=True)
```

```python
final_data.columns.to_list()
```

```python
kmean.fit(final_data)
```

```python
kmean.labels_
```

```python
kmean.cluster_centers_
```

# Evaluation

```python
df['Cluster'] = df['Private']
cluster_map = {'Yes':1, 'No':0}
df['Cluster'] = df['Cluster'].map(cluster_map);
df.head()
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
print(confusion_matrix(df['Cluster'], kmean.labels_))
print('\n')
print(classification_report(df['Cluster'], kmean.labels_))
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np 
```

```python
from sklearn.datasets import make_blobs
```

Let's generate our own DataSet with sklearn

```python
data = make_blobs(n_samples=200, n_features=2, centers=4, cluster_std=1.8,
                 random_state=101)
```

```python
type(data)
```

```python
data[0].shape
```

```python
plt.figure(figsize=(10, 8))
plt.scatter(data[0][:, 0], data[0][:, 1]), plt.figure(figsize=(10, 8))
plt.scatter(data[0][:, 0], data[0][:, 1], c=data[1], cmap='rainbow', edgecolor='black');
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=4)
```

```python
kmeans.fit(data[0])
```

```python
kmeans.cluster_centers_
```

```python
plt.figure(figsize=(10, 8))
plt.scatter(data[0][:, 0], data[0][:, 1], c=data[1], cmap='rainbow', edgecolor='black')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=200)
plt.title('Data with initial labels and cluster\'s centers', fontsize=15)
;
```

```python
plt.figure(figsize=(10, 8))
plt.scatter(data[0][:, 0], data[0][:, 1], c=kmeans.labels_, cmap='rainbow', edgecolor='black')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=200)
plt.title('Data with calculated labels and cluster\'s centers', fontsize=15);
```

Let's now compare them properly!

```python
fig, (ax1, ax2) = plt.subplots(1,2, sharey=True, figsize=(10, 8))

ax1.set_title('K Means')
ax1.scatter(data[0][:, 0], data[0][:, 1], c=kmeans.labels_, cmap='rainbow')

ax2.set_title('Original')
ax2.scatter(data[0][:, 0], data[0][:, 1], c=data[1], cmap='rainbow')
```
